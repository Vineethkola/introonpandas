```python
# prompt: import numpy library ,read this file /content/Salaries.csv , apply functions like head,tail,type attributes like shape,column and also check for null values

import pandas as pd
import numpy as np

# Assuming the file Salaries.csv is in the /content directory
# and the pandas DataFrame 'df' is already loaded as in the previous example.


# Display the first few rows of the DataFrame
print(df.head())

# Display the last few rows of the DataFrame
print(df.tail())


# Get the dimensions (rows, columns) of the DataFrame
print("\nShape of the DataFrame:", df.shape)


# Print the column names of the DataFrame
print("\nColumns of the DataFrame:", df.columns)


# Check for null values in the DataFrame
print("\nNull Values:\n", df.isnull())

# Set display options to show all columns without truncation
pd.set_option('display.max_columns', None)

# Display the sum of null values for each column
print("\nSum of null values per column:\n",df.isnull().sum())
```
